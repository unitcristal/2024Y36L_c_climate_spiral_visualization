```python
import pandas as pd
import matplotlib.pyplot as plt

# 평균 컬럼 추가 (Year 제외한 나머지 열 평균)
df["avg"] = df.iloc[:, 1:].replace("***", float("nan")).astype(float).mean(axis=1)

# 최대값 찾기
max_avg = df["avg"].max()

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(df["Year"], df["avg"], marker="o", linestyle="-", color="b", label="Avg Temp")

# 최대값에 평행한 빨간색 선 추가
plt.axhline(y=max_avg, color="r", linestyle="--", linewidth=1, label=f"Max Avg ({max_avg:.2f})")

# 그래프 스타일 설정
plt.xlabel("Year")
plt.ylabel("Avg Value")
plt.title("Yearly Average Trend")
plt.legend()
plt.grid(True)

# 그래프 출력
plt.show()
```
